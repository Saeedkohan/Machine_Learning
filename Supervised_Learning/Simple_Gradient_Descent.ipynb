```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def J_convex(w):
  return w[0]**2+2*w[1]**2

def grad_J_convex(w):
  return np.array([2*w[0],4*w[1]])
```

```python
def gradient_descent(grad,w,learning_rate=0.1 , n_steps=20):
  w=np.array(w)
  path=[w.copy()]  # for visualize
  for step in range(n_steps):
    grad_w=grad(w)
    w=w-learning_rate *grad_w
    path.append(w.copy())
  return np.array(path)
```

```python
w=[4,4]
path_convex=gradient_descent(grad_J_convex,w,learning_rate=0.1, n_steps=20)

w1_vals=np.linspace(-5,5,400)
w2_vals=np.linspace(-5,5,400)

w1,w2=np.meshgrid(w1_vals,w2_vals)
J_vals=J_convex([w1,w2])
```

```python
plt.figure(figsize=(8,6))
countour_levels=np.logspace(-0.5,3,35)
plt.contour(w1,w2,J_vals,levels=countour_levels,cmap="jet")
path_x ,path_y = path_convex[:,0] , path_convex[:,1]
plt.plot(path_x ,path_y,marker='o',color='red',label='Gradient Descent Path')
plt.title("Graddient Descent on a Covex Function")
```
